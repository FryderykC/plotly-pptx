```python
import plotly.graph_objs as go
import numpy as np
import pandas as pd
```

# Reading csv as dataframe and converting date to datetime

```python
df = pd.read_csv("Binance_ADABTC_d.csv")
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])

fig = go.Figure(data = [go.Candlestick(x=df['date'],
                                       open=df['open'],
                                       high=df['high'],
                                       low=df['low'],
                                       close=df['close'])])

fig.update_layout(template="plotly_dark")
fig.update_layout(yaxis_title= "ADA (Cardano) / BTC (Bitcoin)", xaxis_title = "Date")

fig_name = "my_fig.png"
fig.write_image(fig_name)
fig.show()
```

# Saving plotly fig to powerpoint slide

```python
from pptx import Presentation
from pptx.util import Inches

pr1 = Presentation()
slide1_register = pr1.slide_layouts[0]
slide1 = pr1.slides.add_slide(slide1_register)
title1 = slide1.shapes.title
subtitle1 = slide1.placeholders[1]
title1.text = "Hi Bob!"
subtitle1.text = "Testing Python pptx"

slide2_register = pr1.slide_layouts[1]
slide2 = pr1.slides.add_slide(slide2_register)
title2 = slide2.shapes.title
title2.text = "My ADA BTC Graph"

img_name = fig_name
from_left = Inches(1)
from_top = Inches(1.5)
width = Inches(8)
height = Inches(4)
add_picture = slide2.shapes.add_picture(img_name, from_left, from_top, width, height)

pr1.save("test_ppt.pptx")
```
